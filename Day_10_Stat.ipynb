```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

#count plot / bar plot

```python
df = titanic['sex'].value_counts()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df.index, df.values)
plt.title("count of passenger")
plt.xlabel("Gender")
plt.ylabel("count")
```
