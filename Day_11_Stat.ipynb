```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

#univariate analysis

#histogram

```python
sns.histplot(titanic['age'], bins = 40, kde= True)
plt.title("Age distribution")
```

#kde plot

```python
sns.kdeplot(titanic['age'], shade= True, color = 'r')
plt.title("age distribution")
```

```python
titanic['class'].unique()
```

```python
titanic['class'].value_counts()
```

```python
sns.countplot(titanic['sex'], palette= 'hls')
```

#boxplot

```python
sns.boxplot(x= 'class', y ='age', data = titanic, palette= 'rocket')
```

```python
sns.boxplot(x=titanic['class'], y =titanic['age'], palette= 'rocket', hue = titanic['sex'])
```

```python
sns.boxplot(x=titanic['class'], y =titanic['age'], palette= 'rocket', hue = titanic['survived'])
```

#violin plot

```python
sns.violinplot(x=titanic['class'], y =titanic['age'], palette= 'rocket', hue = titanic['sex'])
```

#bivariate

#scatter - num vs num

```python
sns.scatterplot(x =titanic['age'], y= titanic['fare'], hue =titanic['sex'], palette= 'Set1')
```

#lineplot - numerical - time

```python
sns.lineplot(x='age', y='fare', data = titanic)
```

#bar plot - cat vs numer

```python
sns.barplot(x='class', y='fare', data = titanic)
```
